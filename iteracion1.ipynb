```python
#hazme un cluster de los robos por latitud y longitud y graficalo en un scatter plot
from sklearn.cluster import KMeans
import numpy as np

# Filtrar las columnas de latitud y longitud
X = df[['LATITUD', 'LONGITUD']].values

# Eliminar filas con valores NaN
X = X[~np.isnan(X).any(axis=1)]

# Aplicar KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)

# Agregar la columna de cluster al DataFrame original
df['CLUSTER'] = np.nan
df.loc[~np.isnan(X).any(axis=1), 'CLUSTER'] = kmeans.labels_

# Graficar los clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['LONGITUD'], df['LATITUD'], c=df['CLUSTER'], cmap='viridis', alpha=0.5)
plt.title('Clusters de robos por latitud y longitud')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.colorbar(label='Cluster')
plt.show()
```

## Índice de riesgo de robo

Vamos a crear un índice de riesgo de 0 a 100 que combine la frecuencia de robos, la proporción de robos con violencia y la hora del día. Para la hora, se definen tres franjas:
- 7 a 14
- 14 a 18
- 18 a 23
- 23 a 7

El índice se calculará por franja horaria y día de la semana.

```python
# Definir franjas horarias
import numpy as np

def clasificar_franja(hora):
    if 7 <= hora < 14:
        return '7-14'
    elif 14 <= hora < 18:
        return '14-18'
    elif 18 <= hora < 23:
        return '18-23'
    else:
        return '23-7'

# Crear columna de franja horaria
if 'FRANJA' not in df.columns:
    df['FRANJA'] = df['HORA'].apply(clasificar_franja)
else:
    df['FRANJA'] = df['HORA'].apply(clasificar_franja)

# Filtrar solo las franjas de interés
df_franjas = df[df['FRANJA'].isin(['7-14', '14-18', '18-23', '23-7'])]

# Calcular frecuencia de robos por día y franja
freq = df_franjas.groupby(['DIA DE LA SEMANA', 'FRANJA']).size().rename('FRECUENCIA')

# Calcular proporción de robos con violencia por día y franja
viol = df_franjas.groupby(['DIA DE LA SEMANA', 'FRANJA'])['VOLENCIA'].apply(lambda x: (x == 'SI').mean()).rename('PROPORCION_VIOLENCIA')

# Normalizar frecuencia a 0-1
freq_norm = (freq - freq.min()) / (freq.max() - freq.min())

# Índice de riesgo: 70% frecuencia + 30% violencia
riesgo = 0.7 * freq_norm + 0.3 * viol

# Escalar a 0-100
riesgo_esc = (riesgo - riesgo.min()) / (riesgo.max() - riesgo.min()) * 100

# Unir resultados en un DataFrame
riesgo_df = pd.concat([freq, viol, riesgo_esc.rename('INDICE_RIESGO')], axis=1).reset_index()

riesgo_df.sort_values('INDICE_RIESGO', ascending=False)
```
