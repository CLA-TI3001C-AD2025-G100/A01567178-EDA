```python
import pandas as pd
df = pd.read_excel('data/IPH_robos_ene-ago-2025_tecmty.xlsx')
df.head()
```

```python
# sacar proporción de con violencia o sin violencia de la columna 'VOLENCIA' y graficarla
proportion = df['VOLENCIA'].value_counts(normalize=True)
print(proportion)

import matplotlib.pyplot as plt

proportion.plot(kind='bar')
plt.title('Proporción de robos con y sin violencia')
plt.ylabel('Proporción')
plt.xlabel('Tipo de violencia')
plt.xticks(rotation=0)
plt.show()
```

```python
# ahora para la columna de 'TIPO' sacar el total de cada tipo de robo y si es con violencia o sin violencia
type_proportion = df.groupby('TIPO')['VOLENCIA'].value_counts().unstack()
print(type_proportion)

import matplotlib.pyplot as plt

type_proportion.plot(kind='bar', stacked=True)
plt.title('Proporción de robos con y sin violencia por tipo de robo')
plt.ylabel('Proporción')
plt.xlabel('Tipo de robo')
plt.legend(title='Tipo de violencia')
plt.tight_layout()
plt.show()
```
